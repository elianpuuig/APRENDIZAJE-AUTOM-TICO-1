```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/meatspec.csv")
# df = pd.read_csv("meatspec.csv")
#Elimine la primer columna ya que no aporta informacion relevante al modelo
df = df.drop(columns = df.columns[0])
df
```

1. Preparación: Limpiar columnas innecesarias y dividir el dataset. 📊

2. Línea de base: Entrenar una Regresión Lineal tradicional para comparar. 📉

3. Regularización: Aplicar los tres mosqueteros: Ridge, Lasso y Elastic Net. 🛡️

4. Comparación: Evaluar matemáticamente cuál modelo predice mejor la grasa. 🏆

```python
#1) Preparacion del modelo de regresion lineal

#Preparacion de las variables predictoras X son todos los datos menos fat
X = df.drop(columns=['fat'])

#Preparacion de la variable objetivo Y (fat)
y = df['fat']

#Datos de prueba y test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=123)
```

```python
#2) Linea de base
#Modelo basico de regresion lineal para comparar
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error

modelo_rl = LinearRegression()

modelo_rl.fit(X_train,y_train)

predicciones = modelo_rl.predict(X=X_test)
predicciones = predicciones.flatten()
predicciones[:10]

#Evaluando el modelo de regresion lineal con root_mean_squared_error

rmse_ols = root_mean_squared_error(
            y_true  = y_test,
            y_pred  = predicciones,
           )

print(f"El error (rmse) de test es: {rmse_ols}")

```

**Ridge:** *Ridge le pone un "freno" a los coeficientes de las variables para que no crezcan demasiado y no memoricen el ruido de esas 100 longitudes de onda que están tan correlacionadas.
Digamos que Ridge achica los coeficientes pero los mantiene a todos en la ecuación
*

Al hiperparámetro de penalización $\lambda$ se lo llama alpha

*En esta sección aplicamos el modelo Ridge. Su objetivo es aplicar una penalización que achica los coeficientes de las 100 variables por igual para que ninguna domine. Es ideal aquí porque nuestras variables están muy correlacionadas. Vemos que logra bajar el error de la regresión común de manera drástica.*

```python
#3) Ridge (L2)
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV


alphas = np.logspace(-10,2,200)

modelRidge = RidgeCV(alphas, store_cv_values=True)
modelRidge.fit(X=X_train, y=y_train)

prediccionRidge = modelRidge.predict(X=X_test)
prediccionRidge = prediccionRidge.flatten() #lo convertimos a lista para mejor visualizacion

#Evaluacion de errores
rmse_ridge = root_mean_squared_error(
    y_true=y_test,
    y_pred=prediccionRidge)

print(f"Los Valores de Ridge: {rmse_ridge}")
```

*Acá implementamos Lasso. A diferencia de Ridge, Lasso busca apagar variables a cero para hacer selección de características. Al ejecutarlo con estas variables tan correlacionadas, el modelo sufrió para converger, lo que generó advertencias en la consola y nos obligó a subir las iteraciones máximas a 100.000 para forzar su cálculo.*

```python
#Lasso(L1)
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV

alphas = np.logspace(-5, 3, 200)
#con este me dio un valor de 1
modeloLasso = LassoCV(alphas=alphas, cv=10, random_state=123, max_iter=10000)
#modeloLasso = LassoCV(alphas=alphas ,fit_intercept=True)
modeloLasso.fit(X=X_train, y=y_train)

prediccionLasso = modeloLasso.predict(X=X_test)
prediccionLasso = prediccionLasso.flatten()

#Evaluacion de errores
rmse_lasso = root_mean_squared_error(
    y_true=y_test,
    y_pred=prediccionLasso)

print(f"Variables apagadas a cero Lasso: {np.sum(modeloLasso.coef_ == 0)}")
```

explicacion de lasso

Como el dataset tiene 100 variables (longitudes de onda) que están súper correlacionadas entre sí (se mueven muy juntas).

Ridge maneja esto genial porque su regla es achicar a todas por igual.

Lasso, en cambio, se vuelve loco cuando hay variables casi idénticas. No sabe a cuál de las dos "apagar" y a cuál dejar encendida. El cálculo matemático empieza a rebotar sin llegar a un resultado claro.


Elastic Net es una excelente opcion ya que este combina ambas penalizaciones de Ridge y Lasso en la misma fórmula matemática.

Ridge (L2) maneja genial las variables correlacionadas (como nuestras 100 luces).

Lasso (L1) es buenísimo apagando variables inútiles a cero.

El l1_ratio le dice al modelo qué porcentaje de la mezcla debe ser Lasso. (Si es 1, actúa 100% como Lasso; si es 0, actúa 100% como Ridge). Por eso, en lugar de darle un solo número, le damos una lista de opciones para que el algoritmo pruebe y elija la mejor mezcla.


```python
#Elastic Net
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV

l1_ratio = [0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]
modeloElastic = ElasticNetCV(alphas=alphas,l1_ratio=l1_ratio, cv=10)
modeloElastic.fit(X=X_train,y=y_train)

prediccionElastic = modeloElastic.predict(X_test)
prediccionElastic = prediccionElastic.flatten()

#Evaluacion de errores de Elastic
rmse_elastic = root_mean_squared_error(
    y_true=y_test,
    y_pred=prediccionElastic)

print(f"Evaluacion de Elastic: {rmse_elastic}")
```

# COMPARACION FINAL DE  RIDGE LASSO Y ESLASTIC NET

```python
import plotly.express as px

df_comparacion = pd.DataFrame({
                    'modelo': ['OLS', 'Ridge', 'Lasso', 'Elastic-net'],
                    'test rmse': [rmse_ols, rmse_ridge, rmse_lasso, rmse_elastic]
                 })

df_comparacion = df_comparacion.sort_values(by='test rmse', ascending=True)

graficoComparativo = px.bar(
    df_comparacion,
    x='test rmse',
    y='modelo',
    orientation='h',
    text='test rmse',
    title='Comparación Definitiva de Modelos: Predicción de Grasa 🥩',
    labels={'test rmse': 'Evaluacion de errores RMSE ', 'modelo': 'Modelo'},
    color='test rmse',
    color_continuous_scale='Cividis',
    template='plotly_dark'
)


graficoComparativo.update_traces(texttemplate='%{text:.2f}', textposition='outside')
graficoComparativo.update_layout(
    coloraxis_showscale=False,
    title_font_size=16,
    width=800,
    height=400,
    xaxis=dict(showgrid=True, gridcolor='#333333'),
    yaxis=dict(showgrid=False)
)

graficoComparativo.show()
```
